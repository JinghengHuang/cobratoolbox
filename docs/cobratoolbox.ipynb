```python
import os
import re

# Directory containing the website files
website_directory = 'C:/Users/aaron/OneDrive/Documents/GitHub/cobratoolbox_fork/docs'

# Patterns to search for and replace
patterns = {
    r'/stable/': '/',
    r'/latest/': '/'
}

# Function to find and replace links in files
def find_and_replace_links(directory, patterns):
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith('.html'):  # Only search in HTML files
                file_path = os.path.join(root, file)
                with open(file_path, 'r', encoding='utf-8') as f:
                    content = f.read()
                    
                # Find and replace each pattern
                modified_content = content
                for pattern, replacement in patterns.items():
                    modified_content = re.sub(rf'href=["\']([^"\']*{pattern}[^"\']*)["\']', 
                                              lambda m: m.group(0).replace(pattern, replacement), 
                                              modified_content)
                
                # If changes were made, save the modified content back to the file
                if modified_content != content:
                    with open(file_path, 'w', encoding='utf-8') as f:
                        f.write(modified_content)
                    print(f'Modified {file_path}')

# Run the function
find_and_replace_links(website_directory, patterns)
```

```python
import os
import re

# Directory containing the website files
website_directory = 'C:/Users/aaron/OneDrive/Documents/GitHub/cobratoolbox_fork/docs'

# Old and new base URLs
old_base_url = 'https://opencobra.github.io/cobratoolbox/'
new_base_url = 'https://aaronbrennan1.github.io/cobratoolbox_fork/'

# Function to find and replace the base URL in files
def update_base_url(directory, old_url, new_url):
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith('.html'):  # Only search in HTML files
                file_path = os.path.join(root, file)
                with open(file_path, 'r', encoding='utf-8') as f:
                    content = f.read()
                
                # Replace old base URL with new base URL in href attributes
                modified_content = re.sub(rf'href=["\']{old_url}([^"\']*)["\']', 
                                          lambda m: f'href="{new_url}{m.group(1)}"', 
                                          content)
                
                # If changes were made, save the modified content back to the file
                if modified_content != content:
                    with open(file_path, 'w', encoding='utf-8') as f:
                        f.write(modified_content)
                    print(f'Updated base URL in {file_path}')

# Run the function
update_base_url(website_directory, old_base_url, new_base_url)
```
